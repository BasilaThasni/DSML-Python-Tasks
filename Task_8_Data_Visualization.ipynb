##3. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.
```
Sample Financial data (fdata.csv):
Date,Open,High,Low,Close
10-03-16,774.25,776.065002,769.5,772.559998
10-04-16,776.030029,778.710022,772.890015,776.429993
10-05-16,779.309998,782.070007,775.650024,776.469971
10-06-16,779,780.47998,775.539978,776.859985
10-07-16,779.659973,779.659973,770.75,775.080017
The code snippet gives the output shown in the following screenshot:
```



image.png

````python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame with the provided data (alternatively, load from a CSV file)
data = {
    "Date": ["10-03-16", "10-04-16", "10-05-16", "10-06-16", "10-07-16"],
    "Open": [774.25, 776.030029, 779.309998, 779, 779.659973],
    "High": [776.065002, 778.710022, 782.070007, 780.47998, 779.659973],
    "Low": [769.5, 772.890015, 775.650024, 775.539978, 770.75],
    "Close": [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
}

df = pd.DataFrame(data)

df['Date'] = pd.to_datetime(df['Date'])

plt.plot(df['Date'], df['Open'], label="Open")
plt.plot(df['Date'], df['High'], label="High")
plt.plot(df['Date'], df['Low'], label="Low")
plt.plot(df['Date'], df['Close'], label="Close")

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Alphabet Inc. Stock Prices (Oct 3, 2016 - Oct 7, 2016)')

plt.legend()

plt.show()
````

##Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this.

image.png

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url="https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv"

df = pd.read_csv(url, storage_options={'User-Agent': 'Mozilla/5.0'})

months = df['month_number']
facecream_sales = df['facecream']
facewash_sales = df['facewash']

bar_width = 0.25  # Width of the bars
r1 = np.arange(len(months))  # The position of bars on the x-axis for face cream
r2 = [x + bar_width for x in r1]  # The position of bars on the x-axis for facewash

plt.figure(figsize=(10, 6))
plt.bar(r1, facecream_sales, color='blue', width=bar_width, label='Face Cream Sales Data')
plt.bar(r2, facewash_sales, color='orange', width=bar_width, label='Face Wash Sales Data')

plt.xlabel('Month Number')
plt.ylabel('Sales unit in number')
plt.title('Face wash and Face cream Sales Data')
plt.xticks([r + bar_width / 2 for r in range(len(months))], months)

plt.legend(loc='upper left')
plt.grid(True,linestyle="--")

plt.show()
````

##Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.



image.png

````python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv"

sales_data = pd.read_csv(url,storage_options={'User-Agent':'Mozilla/5.0'})

total_units = {
    'FaceCream': sales_data['facecream'].sum(),
    'FaceWash': sales_data['facewash'].sum(),
    'ToothPaste': sales_data['toothpaste'].sum(),
    'Bathing soap': sales_data['bathingsoap'].sum(),
    'Shampoo': sales_data['shampoo'].sum(),
    'Moisturizer': sales_data['moisturizer'].sum()
}

products = list(total_units.keys())
sales = list(total_units.values())

plt.figure(figsize=(7,7))
wedges, texts, autotexts = plt.pie(sales, labels=products, autopct='%1.1f%%', colors=['lightblue', 'orange', 'green', 'red', 'purple', 'brown'])

plt.legend(wedges, products, title="Products", loc="lower left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Sales Data')

plt.show()
````

##Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot
The Subplot should look like this.

````python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv"

sales_data = pd.read_csv(url,storage_options={'User-agent':'Mozilla/5.0'})

months = sales_data['month_number']
bathingsoap_sales = sales_data['bathingsoap']
facewash_sales = sales_data['facewash']

# Create a figure and subplot structure
plt.figure(figsize=(7,7))

plt.subplot(2, 1, 1)
plt.plot(months, bathingsoap_sales, color='black', marker='o', markeredgewidth='5',linestyle='-', linewidth=5)
plt.title('Sales data of a Bathing soap')
plt.ylabel('Sales units in number')

plt.subplot(2, 1, 2)
plt.plot(months, facewash_sales, color='red', marker='o',markeredgewidth='5', linestyle='-', linewidth=5)
plt.title('Sales data of a Facewash')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')

plt.tight_layout()
plt.show()
````

##Exercise Question 10: Read all product sales data and show it using the stack plot
The Stack plot should look like this.

image.png


````python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv"

sales_data = pd.read_csv(url,storage_options={'User-Agent':'Mozilla/5.0'})

months = sales_data['month_number']
facecream_sales = sales_data['facecream']
facewash_sales = sales_data['facewash']
toothpaste_sales = sales_data['toothpaste']
bathingsoap_sales = sales_data['bathingsoap']
shampoo_sales = sales_data['shampoo']
moisturizer_sales = sales_data['moisturizer']

plt.figure(figsize=(10, 7))
plt.stackplot(months, facecream_sales, facewash_sales, toothpaste_sales, bathingsoap_sales, shampoo_sales, moisturizer_sales,
              labels=['FaceCream', 'FaceWash', 'ToothPaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'],
              colors=['purple', 'cyan', 'red', 'black', 'green', 'yellow'])

plt.legend(loc='upper left')
plt.title('All products sales data using stack plot')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')

plt.show()
````
